```python
#import necessary packages and display the raw data
import pandas as pd
import numpy as np
df = pd.read_csv('adult_income.csv')
df
```

```python
df.info()
```

```python
df['income']=df[' Income'].map(lambda x: 0 if x ==' <=50K' else 1)
print(df.shape)
```

```python
df.drop_duplicates(keep='first',inplace=True)
print(df.shape)
```

```python
df[(df[' Work-class']==' ?')|(df[' Occupation']==' ?') |(df[' Native-country']==' ?')].shape
```

```python
df=df.drop(index=df[(df[' Work-class']==' ?')|(df[' Occupation']==' ?') |(df[' Native-country']==' ?')].index)
print(df.shape)
```

```python
#1st dummy variable - Work Class
dummies_1 = pd.get_dummies(df[' Work-class'],drop_first = False)
dummies_1
```

```python
#2nd dummy variable -  Marital-status
dummies_2 = pd.get_dummies(df[' Marital-status'],drop_first = False)
dummies_2
```

```python
#3rd dummy variable -  Occupation
dummies_3 = pd.get_dummies(df[' Occupation'],drop_first = False)
dummies_3
```

```python
#4th dummy variable -  Relationship
dummies_4 = pd.get_dummies(df[' Relationship'],drop_first = False)
dummies_4
```

```python
#5th dummy variable -  Race
dummies_5 = pd.get_dummies(df[' Race'],drop_first = False)
dummies_5
```

```python
#6th dummy variable -  Sex
dummies_6 = pd.get_dummies(df[' Sex'],drop_first = False)
dummies_6
```

```python
df = pd.concat([df,dummies_1,dummies_2,dummies_3,dummies_4,dummies_5,dummies_6],axis = 1)
df.drop([' Work-class',' Marital-status',' Occupation',' Relationship',' Race',' Sex'], axis=1, inplace=True)
df
```

```python
df[' Native-country']=df[' Native-country'].map(lambda x: 0 if x ==' United-States' else 1)
```

```python
# Eduction and Eduction_num are essentially duplicate
df.drop(' Education', axis=1, inplace=True)
```

```python
df['Capital'] = df[' Capital-gain'] + df[' Capital-loss']
```

```python
df.drop(' Capital-gain', axis=1, inplace=True)
df.drop(' Capital-loss', axis=1, inplace=True)
df.info()
```

```python
df.drop(' Income', axis=1, inplace=True)
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df[[' fnlwgt','Capital',' Hours-per-week']])
df[[' fnlwgt','Capital',' Hours-per-week']] = scaler.transform(df[[' fnlwgt','Capital',' Hours-per-week']])
df
```

```python
from sklearn.model_selection import KFold, cross_val_score
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
import sklearn.metrics as metrics
from sklearn.model_selection import train_test_split

x = df.drop(['income'], axis=1)
y = df['income']

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2)

other_params = {'eta': 0.3, 'n_estimators': 500, 'gamma': 0, 'max_depth': 6, 'min_child_weight': 1,
                'colsample_bytree': 1, 'colsample_bylevel': 1, 'subsample': 1, 'reg_lambda': 1, 'reg_alpha': 0,
                'seed': 33}
cv_params = {'n_estimators': np.linspace(100, 1000, 10, dtype=int)}
regress_model = xgb.XGBClassifier(**other_params)
gs = GridSearchCV(regress_model, cv_params, verbose=2, refit=True, cv=5, n_jobs=-1)

gs.fit(x_train, y_train)

y_pred = gs.predict(x_train)

#Performance
print("ROC_AUC:", metrics.roc_auc_score(y_train,y_pred))
print("Best value for parametres:", gs.best_params_)
print("Best score for model:", gs.best_score_)
```
